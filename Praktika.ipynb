```python
def get_median(a):
    #a=[5,2,1,3,4]
    #a=[3, 3, 7, 9]
    por=sorted(a)
    if len(a)%2!=0:
        poza=int(len(a)/2-0.5)
        rezult=por[poza]
    else:
        poza=int(len(a)/2-1)
        pozb=int(len(a)/2)
        rezult=int((por[poza]+por[pozb])/2)
    return rezult
print(get_median([3, 3, 7, 9]))
```

```python
def to_euro(price):  
    exchange_rate = 70  
    rounded = round(price/exchange_rate, 2)  
    return '€' + str(rounded)  
  
def db_to_euro(db):  
    return [to_euro(item['price']) for item in db]  
  
  
ticket_db = [{'price': 400}, {'price': 200}, {'price': 150}]  
guide_db = [{'price': 50}, {'price': 40}]  
snack_db = [{'price': 100}, {'price': 95}, {'price': 150}]  
  
tickets_euro = db_to_euro(ticket_db)  
guides_euro = db_to_euro(guide_db)  
snacks_euro = db_to_euro(snack_db)  
  
print(tickets_euro, guides_euro, snacks_euro)  
# => ['€5.71', '€2.86', '€2.14'] ['€0.71', '€0.57'] ['€1.43', '€1.36', '€2.14']  
```

```python
# добавьте функцию get_euro_rate
def get_euro_rate():
    from random import random
    exchange_rate=random()*(85-65)+65
    return exchange_rate
# используйте get_euro_rate в следующей функции
def to_euro(price):  
    exchange_rate = get_euro_rate()
    rounded = round(price/exchange_rate, 2)  
    return '€' + str(rounded)
print(to_euro(100))
```

```python
values = [-7, -7, 7, 7]
std = 42

#этот кусок вообще можно было не считать в уме для задания
def count_std():
    mean = sum(values)/len(values)
    std = (sum([(value-mean)**2 for value in values])/len(values))**0.5
    return std
print(count_std())

def normalize(value):
    result = value/std
    return result 

print(normalize(21))
```

```python
def normalize(numbers,mean=0,std=1):
    return list(map(lambda x: int((x-mean)/std),numbers))
normalize([10, 20])
normalize([10, 20], std=2)
normalize([10, 20], mean=15)
```

```python
def normalize(numbers,mean=0,std=1):
    result=[]
    for x in numbers:
        result.append(int((x-mean)/std))
    return result
normalize([10, 20])
normalize([10, 20], std=2)
normalize([10, 20], mean=15)
```

```python
#Определите функцию count_letters(sentence, average), которая считает количество букв в строке без учёта пробелов
def count_letters(sentence, average=False):
    info=[]
    info=sentence.split()
    count_words=len(info)
    letters=sum(len(a) for a in info)
    #print(info)
    #print('количество слов: ',count_words)
    #print(letters)
    if average == True:
        result= letters/count_words
    else:
        result=letters
    return result
#count_letters('что здесь написано?')
#count_letters("Beep boop")
#count_letters("Beep boop", average=True)
#count_letters("I will build my own theme park")
count_letters("I will build my own theme park", average=True)
```
