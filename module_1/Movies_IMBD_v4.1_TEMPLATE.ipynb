# Предобработка

```python
answers = {} # создадим словарь для ответов

# загружаем массив данных
df = pd.read_csv('movie_bd_v5.csv')

# глаза разбегаются от количества нулей в цифрах, мы сократим их на 6 нулей и будем считать в миллионах
df['budget']=df['budget']/1000000
df['revenue']=df['revenue']/1000000

# добавим новую колонку с отображением доходности фильма
df['profit']=df['revenue']-df['budget']

#колонку с датой выхода фильма необходимо перевести в формат даты
df['release_date'] = pd.to_datetime(df['release_date'])

# добавим колонку с обозначением месяца выхода фильма
df['month'] = df['release_date'].dt.month

# добавим колонку с длиной названий фильмов
df['name_len'] = df.original_title.apply(lambda x: len(x))

# добавим колонку с количеством слов в описании
df['overview_len'] = df.overview.apply(lambda x: len(x.split(' ')))

# давайте посмотрим на предобработанные данные
display(df.sample(5))
```

# 11. Какого жанра фильмов больше всего?

```python
# эту задачу тоже можно решать разными подходами, попробуй реализовать разные варианты
# если будешь добавлять функцию - выноси ее в предобработку что в начале

answers['11'] = 'Drama'

# определим какие жанры встречаются в выборке
#display(df.genres.unique)
# мы видим, что фильм может относиться к нескольким жанрам одновременно, нам требуется определить исчерпывающий список жанров, встречающихся в выборке. 
# Возможно какой-то жанр не встречается в одиночку и всегда перемешан с другими, поэтому выделим каждый встречающийся жанр из всего перечня
# отделим жанры друг от друга через разделитель '|' и отобразим количество повторений в выборке
display(Counter(df['genres'].str.split('|').sum()))
# наиболее часто встречаемый жанр
display(Counter(df['genres'].str.split('|').sum()).most_common(1))
```

ВАРИАНТ 2

```python
# посмотрим какие жанры встречаются в выборке
genres = []
for i in df['genres']:
# создадим список жанров, который будем дополнять новыми найденными значениями
    genre_list=[]
    genre_list = i.split('|')
    for j in genre_list:
        if j in genres:
            continue
        else:
            genres.append(j)
print('Всего жанров: ',len(genres),genres)

print('Посчитаем количество фильмов, соответствующих каждому из жанров:')
for i in genres:
    count = 0
    for j in df['genres']:
        if i in j:
            count += 1
    print(i, count)
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = 'Daniel Radcliffe & Rupert Grint'
duo = Counter()
for i in range(0,len(df)):
    cinema_stars = df.cast[i].split('|')
    for j in list(itertools.combinations(cinema_stars, 2)):
        if j not in duo:
            duo[j] = 1
        else:
            duo[j] += 1
duo.most_common(3)
```
